# Marble Microblaze Hello

##  1. Devices

- LBNL Marble Board, in LLRF chassis**

## 2. Firmware and software
- Vivado and Vitis
- An Xilinx virtual cable (XVC) is required to connect the Marble board to the Xilinx tool chain (hw_server). [https://github.com/BerkeleyLab/XVC-FTDI-JTAG](https://github.com/BerkeleyLab/XVC-FTDI-JTAG)

**c. Software setup**
            
    i. Create Vitis platform

    ii. Create app using hello world template
